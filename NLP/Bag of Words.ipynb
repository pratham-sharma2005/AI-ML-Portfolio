```python
import pandas as pd

df = pd.DataFrame({"text":[ "people watch BITS" , "BITS watch BITS" , "people write comment" , "BITS write comment" ] , "output":[1,1,0,0]})

df
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv =CountVectorizer()

bow = cv.fit_transform(df["text"])

print(cv.vocabulary_) # number shown are the index where the frequency of the numbers are shown
```

```python
print(bow[0].toarray())
```

```python
x = cv.transform(["people watch and write  and comment BITS but I don't "]) # extra words written are ignored

print(x.toarray())
```

```python
cv1 = CountVectorizer(ngram_range=(2,2)) # MAKES BIGRAM WHICH IS every vocab is made of 2 words 
bow1 = cv1.fit_transform(df["text"])

print(cv1.vocabulary_) # number shown are the index where the frequency of the numbers are shown
```

```python
print(bow1[0].toarray())
```

```python
cv2 = CountVectorizer(ngram_range=(1,2)) # MAKES BIGRAM and UNIGRAM WHICH IS every vocab is made of 2 words and 1 word too
bow2 = cv2.fit_transform(df["text"])

print(cv2.vocabulary_) # number shown are the index where the frequency of the numbers are shown

print(bow2[0].toarray())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf =TfidfVectorizer()

tfidf = TfidfVectorizer() 
check = tfidf.fit_transform(df["text"]).toarray()


print(check)
```
