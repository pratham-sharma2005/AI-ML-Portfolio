```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.preprocessing import PolynomialFeatures
from sklearn.datasets import make_regression
```

```python
x,y = make_regression(n_samples=100, n_features=1 , n_targets= 1 ,noise= 50 , random_state=10)
plt.plot(x,y,"b.")
```

```python
x_train , x_test ,y_train ,y_test = train_test_split(x,y,random_state=42,test_size=0.2)
lr = LinearRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
print("R2_Score for LR:",r2_score(y_test,y_pred))
```

```python
poly = PolynomialFeatures( degree = 2)
x_ploy_train = poly.fit_transform(x_train)
x_ploy_test = poly.transform(x_test)

lr.fit(x_ploy_train,y_train)
y_predict_poly = lr.predict(x_ploy_test)
print("R2 SCORE after polynomial features and applying Lr:",r2_score(y_test,y_predict_poly))
```

```python
L = Lasso(alpha = 0.1 )
L.fit(x_ploy_train,y_train)

y_predict_Lasso = L.predict(x_ploy_test)
print("R2 score by lasso:",r2_score(y_test,y_predict_Lasso))
```

```python
L = Lasso(alpha = 0.1 )
L.fit(x_train,y_train)

y_predict_Lasso_2 = L.predict(x_test)
print("R2 score by lasso:",r2_score(y_test,y_predict_Lasso_2))
```
