```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
df = pd.read_csv("/Users/prathamsharma/Desktop/datasets/DT_KNN/spambase_data_2.csv",header=None)
```

```python
df
```

```python
x= df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
x_train, x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=30)
```

```python
tree = DecisionTreeClassifier()
tree.fit(x_train,y_train)
y_pred = tree.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.tree import plot_tree
plot_tree(tree)
plt.show()
```

```python
stump = DecisionTreeClassifier(max_depth=1,random_state=30)
stump.fit(x_train,y_train)
y_pred_stump = stump.predict(x_test)
stump_accuracy = accuracy_score(y_test,y_pred_stump)
stump_error = 1 - stump_accuracy
print(stump_accuracy)
plot_tree(stump)
```

```python
tree_entropy = DecisionTreeClassifier(criterion="entropy",random_state=30)
tree_entropy.fit(x_train,y_train)
y_pred_tree_entropy = tree_entropy.predict(x_test)
tree_entropy_accuracy = accuracy_score(y_test,y_pred_stump)
tree_entropy_error = 1 - tree_entropy_accuracy
plot_tree(tree_entropy)
plt.show()
print(tree.get_depth())
```

```python
max_depth_accuracies = []
for depth in range(1, 21):  # Try depths from 1 to 20
    model = DecisionTreeClassifier(criterion='entropy', max_depth=depth, random_state=30)
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    acc = accuracy_score(y_test, y_pred)
    max_depth_accuracies.append((depth, acc))

# Find depth with best accuracy
best_depth, best_accuracy = max(max_depth_accuracies, key=lambda x: x[1])
print(f"Best Accuracy: {best_accuracy:.4f} at Depth: {best_depth}")
```

```python
import heapq
# Get feature importances
feature_imp = tree_entropy.feature_importances_

# Get indices of top 10 features
top_10_indices = heapq.nlargest(10, range(len(feature_imp)), key=feature_imp.__getitem__)
top_10_features = x.columns[top_10_indices]

print("Top 10 important features:")
print(top_10_features)

# Build model using top 10 features
X_train_top10 = x_train[top_10_features]
X_test_top10 = x_test[top_10_features]

model_top10 = DecisionTreeClassifier(criterion='entropy', random_state=30)
model_top10.fit(X_train_top10, y_train)
y_pred_top10 = model_top10.predict(X_test_top10)

top10_accuracy = accuracy_score(y_test, y_pred_top10)
print(f"Accuracy using Top 10 Features: {top10_accuracy:.4f}")

# Check for information loss
if abs(tree_entropy_accuracy - top10_accuracy) < 0.01:
    print("Minimal or no information loss.")
elif top10_accuracy < tree_entropy_accuracy:
    print("Some information loss occurred.")
else:
    print("Unexpected: Top 10 features performed better.")
```
